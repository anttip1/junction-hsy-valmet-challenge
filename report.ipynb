```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
algorithm_df = pd.read_csv("simulation_output_0_58_27.csv")
```

```python
benchmark_df = pd.read_csv("Hackathon_HSY_data.csv")
```

```python

time_stamps = benchmark_df["Time stamp"]
water_level_column = "Water level in tunnel L1 (m)"

plt.figure(figsize=(12, 6))
plt.plot(time_stamps, benchmark_df[water_level_column], label="Water Level L1 (m) benchmark")
plt.plot(time_stamps, algorithm_df[water_level_column], label="Water Level L1 (m) algorithm")
plt.xlabel("Time")
plt.ylabel("Water Level (m)")
plt.title("Water Level Timeseries")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Outflow plot
plt.figure(figsize=(12, 6))
plt.plot(time_stamps, benchmark_df["Sum of pumped flow to WWTP F2 (m3/h)"]/4, label="Outflow from tunnel (m3/15 min) benchmark")
plt.plot(time_stamps, algorithm_df[ "Outflow (m3/15 min)"], label="Outflow from tunnel (m3/15 min) algorithm")
plt.xlabel("Time")
plt.ylabel("Outflow (m3/15 min)")
plt.title("Outflow Timeseries")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
